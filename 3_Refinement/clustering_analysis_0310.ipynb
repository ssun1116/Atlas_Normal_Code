```python
import pandas as pd
import anndata

# Assuming you have an AnnData object named 'adata'

# Step 1: Group the data by Leiden clusters
grouped = adata.obs.groupby('Leiden')

# Step 2: Iterate over each group
for cluster, group_data in grouped:
    print(f"Cluster {cluster}:")
    
    # Step 3: Count occurrences of each brain region within the current cluster
    region_counts = group_data['Brain Region'].value_counts(normalize=True)
    
    # Step 4: Sort the counts and print only the top 3 brain regions
    top3_regions = region_counts.head(3)
    
    # Step 5: Print the proportion of each brain region within the cluster
    print(top3_regions)
    print()
```
